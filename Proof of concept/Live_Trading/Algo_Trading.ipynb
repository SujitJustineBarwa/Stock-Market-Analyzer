```python
# Get Historical Data
import datetime
instrument_token = 779521
from_datetime = datetime.datetime.now() - datetime.timedelta(days=30)     # From last & days
to_datetime = datetime.datetime.now()
interval = "minute"

historical_data = kite.historical_data(instrument_token, from_datetime, to_datetime, interval, continuous=False, oi=False)
df = pd.DataFrame(historical_data) 
df
```

```python
df['return'] = -df.close.diff(-1)
df['lag_1'] = df['return'].shift()
df['lag_2'] = df['return'].shift(2)
df
```

```python
import matplotlib.pyplot as plt
```

```python
# Drop rows with NaN values in 'return', 'lag_1', or 'volume' for plotting
df.dropna(subset=['return', 'lag_1', 'volume'], inplace=True)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = df['return']
y = df['lag_1']
z = df['volume']

ax.scatter(x, y, z, c='b', marker='o')

ax.set_xlabel('Return')
ax.set_ylabel('Lag_1')
ax.set_zlabel('Volume')

plt.show()
```
